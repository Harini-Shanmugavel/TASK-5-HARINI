```python
# TASK 5 - EDA on Titanic Dataset

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# Load Dataset
titanic = pd.read_csv('/content/Titanic-Dataset.csv')

# Preview Dataset
titanic.head()

# --- Basic Info ---

print("Shape of dataset:", titanic.shape)
titanic.info()
titanic.isnull().sum()


# 1. Violin Plot - Age Distribution by Survival
plt.figure(figsize=(10,6))
sns.violinplot(x='Survived', y='Age', data=titanic, palette='muted')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

print("Observation: Survivors tended to be younger, non-survivors have broader age spread.")

# 2. Swarm Plot - Sex vs Age by Survival
plt.figure(figsize=(10,6))
sns.swarmplot(x='Sex', y='Age', hue='Survived', data=titanic, palette='deep')
plt.title('Age and Sex Distribution by Survival')
plt.show()

print("Observation: Female survivors are concentrated in younger and middle ages.")

# 3. Boxen Plot - Fare Distribution by Passenger Class
plt.figure(figsize=(10,6))
sns.boxenplot(x='Pclass', y='Fare', data=titanic, palette='coolwarm')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

print("Observation: 1st class passengers paid significantly higher fares.")

# --- Bivariate Analysis ---

# 4. Countplot - Embarkation Port vs Survival
plt.figure(figsize=(10,6))
sns.countplot(x='Embarked', hue='Survived', data=titanic, palette='pastel')
plt.title('Survival Counts by Embarkation Port')
plt.show()

print("Observation: Passengers from 'C' (Cherbourg) had better survival rates.")

# 5. Catplot - Pclass vs Survived by Gender
sns.catplot(x='Pclass', hue='Sex', col='Survived', kind='count', data=titanic, palette='Set2')
plt.suptitle('Passenger Class and Survival by Gender', y=1.05)
plt.show()

print("Observation: Higher fare and lower passenger class (1st class) passengers had higher survival rates.")
```

```python
# 6. Jointplot - Age vs Fare by Survival
sns.jointplot(data=titanic, x='Age', y='Fare', hue='Survived', kind='scatter', palette='husl')
plt.suptitle('Joint Distribution of Age and Fare by Survival', y=1.02)
plt.show()
```

```python
# 7. Correlation Heatmap with Mask
corr = titanic.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(12,8))
sns.heatmap(corr, mask=mask, cmap='YlGnBu', annot=True, linewidths=0.5)
plt.title('Correlation Matrix (Upper Triangle Masked)')
plt.show()
```
